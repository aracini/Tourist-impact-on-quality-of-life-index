```python

# Ruta a los archivos SHP
ruta_completa_shp_1_2 = 'Objetivo 1/shp_indice_pobreza_multidimen/indice_pobreza_multidimen.shp'
ruta_completa_shp_4_1 = 'Objetivo 4/shp_sedes_educativas 2023/sedes_educativas.shp'

try:
    # Cargar el SHP de "Objetivo 1" como fondo
    gdf_fondo = gpd.read_file(ruta_completa_shp_1_2)

    # Cargar el SHP de "Objetivo 4" (datos de sedes educativas)
    gdf_sedes = gpd.read_file(ruta_completa_shp_4_1)

    # Asegurarse de que ambos están en el mismo sistema de coordenadas
    gdf_sedes = gdf_sedes.to_crs(gdf_fondo.crs)

    # Crear el gráfico
    fig, ax = plt.subplots(figsize=(10, 8))

    # Graficar el fondo (Mapa de Medellín)
    gdf_fondo.plot(ax=ax, color="lightgrey", edgecolor="black")

    # Graficar las sedes educativas sobre el fondo
    gdf_sedes.plot(ax=ax, color="blue", marker='o', markersize=5, alpha=0.7)

    plt.title("Mapa de Medellín con Sedes Educativas - 2023")
    plt.xlabel("Longitud")
    plt.ylabel("Latitud")
    plt.show()

except Exception as e:
    print(f"Error al cargar los archivos SHP: {e}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
csv_path = 'Objetivo 4/ISDCE/indice_sintetico_de_calidad_educativa 2015-2018.csv'
df_seleccionado = pd.read_csv(csv_path)[['comuna', 'isce_total_2018']]

ruta_comunas_csv = 'Objetivo 1/csv_indice_desarrollo_humano 2014-2021/indice_desarrollo_humano.csv'
df_comunas = pd.read_csv(ruta_comunas_csv, encoding='latin1')

# Realizar la unión de los DataFrames
df_unido = df_comunas.merge(df_seleccionado, left_on='codigo', right_on='comuna', how='inner')

# Agrupar por comuna y calcular el valor promedio de 'isce_total_2018'
df_promedio_comunas = df_unido.groupby('comuna')['isce_total_2018'].mean().reset_index()

# Preparar el gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(df_promedio_comunas['comuna'], df_promedio_comunas['isce_total_2018'], 
            s=df_promedio_comunas['isce_total_2018']*10,  # Ajustar el tamaño de los puntos
            c=df_promedio_comunas['isce_total_2018'], cmap='viridis', alpha=0.6)

# Etiquetas y título
plt.title('Gráfico de Dispersión por Comuna con Promedio de ISDCE')
plt.xlabel('Comuna')
plt.ylabel('Promedio ISDCE Total 2018')

# Mejorar la visualización del gráfico
plt.xticks(rotation=90)  # Rotar etiquetas del eje X para mejor lectura
plt.colorbar(label='ISDCE Total 2018')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
